# Beam splitters and interferometers

Most of the sources we encounter in quantum optics are not single photon sources, specially because of the probability of emitting more than one photon at a time. Experimentally, such task is very difficult to achieve and was only recently demonstrated using bounced photons irradiated from a double decaying calcium atom, and detecting them requires a beam splitter. In this section we will discuss the basic properties of beam splitters and how they can be used to build interferometers.

## Beam splitters

In comparisson to the classical case, studying a BS requires special treatment to make sure we don't miss the conmmutation relation of the input and output operators. 



This can be seen on the meantime like a black box that takes two input modes and outputs two modes. The input and output operators are related by the following transformation:

$$\begin{pmatrix}
\hat{a}_2 \\
\hat{a}_3
\end{pmatrix} = \begin{pmatrix}
t' & r \\
r' & t
\end{pmatrix} \begin{pmatrix}
\hat{a}_0 \\
\hat{a}_1
\end{pmatrix}$$

where $t$ and $r$ are the transmission and reflection coefficients which have to satisfy energy conservation:

$$|t|^2 + |r|^2 = 1, \quad r^*t' + t^*r' = 0, \quad r^*t + r't'^* = 0$$

For example, for a 50/50 beam splitter, we have $t = r = 1/\sqrt{2}$ and $t' = -r' = i/\sqrt{2}$ such that the input and output operators are related by:

$$\hat a_2 = \frac{1}{\sqrt{2}}(\hat a_0 + i \hat a_1), \quad \hat a_3 = \frac{1}{\sqrt{2}}(\hat a_0 + i \hat a_1)$$

We can also consider for example a single photon input state $\ket{1,0}$ and see what happens to the output state:

$$\ket{1,0} \rightarrow \frac{1}{i\sqrt{2}}(\ket{1,0} + \ket{0,1})$$

The last mapping yields an interesting conclusion: if only one photon is detected at the output, we can't tell which input mode it came from.

Lastly, we can also consider the case of a coherent state input $\ket{0, \alpha}$ and see what happens to the output state. Note that a coherent state is reachable quite easily using a displaced vacuum state $\ket{\alpha} = \hat D(\alpha) \ket{0}$, where $\hat D(\alpha) = e^{\alpha \hat a^\dagger - \alpha^* \hat a}$ is the displacement operator. The output state is then given by:

$$\ket{0, \alpha} \rightarrow \frac{1}{2}\ket{i\alpha, \alpha}$$

Note that this result is only direct for a single photon input state, but we can use the superposition principle to extend it to any input state. For example, for a vacuum + coherent state input $\ket{0}\ket{\alpha}$, we get:

$$\ket{0}\ket{\alpha} \rightarrow e^{-|\alpha|^2/2}\sum_{n=0}^\infty \sum_{m=0}^\infty \frac{(i\alpha/\sqrt{2})^n (\alpha\sqrt{2})^m}{\sqrt{n!m!}} (\ket{n, m})$$
$$= e^{-|\alpha|^2/2}\sum_{n=0}^\infty \sum_{m=0}^\infty \frac{\alpha^N}{N!} \frac{1}{2^{N/2}}\sum_n^N \binom{N}{n} i^{N-2n} (\ket{n, N-n})$$

where $N = n+m$. The last expression can be simplified using the binomial. The result is:

$$\ket{0}\ket{\alpha} \rightarrow e^{-|\alpha|^2/2}\sum_{N=0}^\infty \frac{\alpha^N}{N!} \sum_n^N \binom{N}{n} i^{N-2n} (\ket{n, N-n})$$

## Interferometers with single $\gamma$

Let us consider a Mach–Zehnder interferometer (MZI), which is simply a combination of two beam splitters and two mirrors, where one of the paths introduces a phase shift $\hat S = e^{i\theta}$:

[Gerry at al.]

If we assume a single photon comming in ($\ket{0, 1}$), we can calculate the output state using:

$$\ket{0, 1} \rightarrow \frac{1}{\sqrt{2}}(i\ket{1, 0} + \ket{0, 1}) $$

And after the shifter we get:

$$\frac{1}{\sqrt{2}}(\ket{1, 0} + i\ket{0, 1}) \rightarrow \frac{1}{\sqrt{2}}(i\ket{1, 0} + e^{i\theta}\ket{0, 1}) $$

Finally, the second BS acts on both states, yielding:

$$\frac{1}{2}(i\ket{1, 0} + e^{i\theta}\ket{0, 1} + i\ket{0, 1} + e^{i\theta}\ket{1, 0})$$

Thus, the probability of detecting a photon at the output at D1 (D2) is given by $P_+$ ($P_-$):

$$P_\pm = \frac{1}{2}(1 \pm \cos(\theta))$$

Obviously, because of the random behavior of the photon, we can't predict which detector will detect the photon with certainty, but other sources of uncertainty arise from the phase shifter. A phase uncertainty is added given by:

$$\Delta \theta = \frac{1}{\sqrt{\bar n}|\sin(\theta)}$$

Experimentally this is quite nice, and somewhat expected, if we do measurements for a large number of photons ($\bar n = \alpha^2$), the uncertainty drops. In that direction, the bigger the phase shift, the smaller the uncertainty as well.
